# 파이썬을 사용하여 간단한 계산기 프로그램 만들기

---


사용자로부터 두개의 정수와 연산자(+, -, *, /, **)를 입력받아 계산 결과를 출력하는 프로그램 

[조건]
- 사용자가 입력한 값이 '정수'가 아닌 경우 ValueError 처리
- 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.” 출력 + 다시 입력받기 시도
- 나눗셈 연산 시, num2가 0 일 경우, ZeroDivisionError
- math 모듈을 사용하여 제곱 연산 지원 (math.pow)
- 계산기를 통해 계속 계산할 것인지 입력 받는다


```python
# 계산 과정을 수행할 calculator 함수 정의 
def calculator():
    import math as m

    # while True 문을 사용하여 반복해서 계산 가능하게 설계
    while True:
        try:
            # 인수를 받아서
            user_input = input("두 정수와 연산자를 차례로 입력해주세요 (예시: 5 6 /)")
            input_list = user_input.split()

            # 사용자가 인수를 제대로 입력했는지 확인
            # [Error] 인수를 제대로 입력하지 않았을 경우 에러 발생. 이를 방지하기 위해서 코드 추가 
            if len(input_list) != 3:
                print("잘못된 입력입니다. '숫자 숫자 연산자' 형식으로 입력해주세요.")
                continue

            # 정수와 연산자를 각각의 변수에 할당 
            num1, num2, op = input_list


            # 정수 변환 (정수 변환이 안 될 경우 except ValueError로 이동)
            user_num1 = int(num1)
            user_num2 = int(num2)

            # 연산자에 따라서 연산 수행
            if op == '+':
                result = user_num1 + user_num2
            elif op == '-':
                result = user_num1 - user_num2
            elif op == '*':
                result = user_num1 * user_num2
            elif op == '/':
                if user_num2 == 0:
                    raise ZeroDivisionError('0으로는 나눌 수 없습니다.')
                else:
                    result = user_num1 / user_num2
            elif op == "**":
                result = m.pow(user_num1, user_num2)
            else:
                print("지원하지 않는 연산자입니다. (+, -, *, /, ** 중 입력)")
                continue

            # 결과 출력
            print(f"결과 :{result}")

            # 계산을 계속할지 여부 확인
            if input("계속 계산하시겠습니까? (yes = 1, no = 0)") == '0':
                print("계산기를 종료합니다.")
                break

        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")
        except ZeroDivisionError as e:
            print(e)
            
```

```python
calculator()
```

## 추가 연습(페어프로그래밍 이후 개인적인 연습)

### [질문] 연산 수행 과정을 줄일 수 없을까? 

---

- operator 모듈을 사용해서, 허용할 연산자를 미리 딕셔너리로 만들어놓는다면?
  - 이 딕셔너리를 이용해서 코드를 줄일 수 있을 것 같다. 

```python
def calculator2():
    # 함수를 독립적인 모듈로 작동시키기 위해서 함수 내부 임포트
    import math as m
    import operator

    # 연산자를 딕셔너리로 구조화
    ops = {
        '+' : operator.add,
        '-' : operator.sub,
        '*' : operator.mul,
        '/' : operator.truediv,
        '**' :operator.pow
    }

    while True:
        try:
            user_input = input("두 개의 정수와 연산자를 공백으로 구분하여 차례로 입력하세요(예: 5 3 +)")
            input_list = user_input.split()

            # 입력값이 3개가 아닌 경우
            if len(input_list) != 3:
                print("잘못된 입력입니다. '숫자 숫자 연산자' 형식으로 입력해주세요.")
                continue    # 입력값이 3개가 아니면 밑의 코드 건너뛰고 처음으로 돌아감

            num1, num2, op = input_list

            # 입력받은 값 정수화
            user_num1 = int(num1)
            user_num2 = int(num2)

            # 지원하는 연산자인지 확인
            if op not in ops:
                print("지원하지 않는 연산자입니다. (+, -, *, /, ** 중 입력)")
                continue   # 처음으로 다시 돌아감

            # 나눗셈인 경우 0으로 나누면 ZeroDivisionError 발생
            if op == '/' and user_num2 == 0:
                raise ZeroDivisionError("0으로 나눌 수 없습니다. ")

            # 정상적인 입력값인 경우 연산 실행
            result = ops[op](user_num1, user_num2)
            print(f"연산 결과 : {result}")

            if input("계속 계산하시겠습니까? (yes = 1, no = 0)") == '0':
                print("계산기를 종료합니다.")
                break

        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")
        except ZeroDivisionError as e:
            print(e)
```

# NOTE

----

생각을 넓히면 코드를 여러개로 더 만들어볼 수 있는 문제였던 것 같은데, 아무래도 페어프로그래밍으로 서로 말로 설명하고 이해하는 시간이 필요하다보니 코드를 여러개 만들 수는 없었다. 그래도 페어와 함께 에러가 발생했을 때 대처할 수 있는 구문을 추가하고, if-elif-else 구문으로 연산을 쪼개서 실행시켜 제대로 코드가 굴러가도록 만들 수 있었다.     
그리고 페어프로그래밍 외에 개인적으로 파이썬 독에서 operator 라는 모듈을 찾아 코드를 줄여보려고 시도해보았다.
