## 퀘스트 1) 은행계좌 만들기

---


````python
import random
from datetime import datetime

class Account:
    account_num = 0    #계좌 개수를 저장할 클래스 속성 지정 

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank_name = 'SC은행'

        # [repact] 계좌번호 : 3자리-2자리-6자리 랜덤 생성
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0,999999)       
        self.account = f"{num1:03d}-{num2:02d}-{num3:06d}"  
        
        self.deposit_num = 0        # 이자 지급을 위해 입금 횟수를 저장할 속성 지정 
        self.deposit_list = []      # 입금 내역 저장할 리스트 
        self.withdraw_list = []     # 출금 내역 저장할 리스트 
        Account.account_num += 1    # 계좌 개설 시 +1

    # 계좌 개수 출력 메서드
    @classmethod             # 계좌를 만든 사용자가 없더라도 계좌 개수 출력이 가능하도록 클래스 메서드로 정의 
    def get_account_num(cls):
        print(f"개설된 계좌의 개수 : {cls.account_num}")

    # 입금 메서드
    def deposit(self, amount):
        # 입금액이 1원 이하인 경우 예외 처리 
        if amount < 1:
            raise ValueError('오류: 최소 1원 이상만 입금할 수 있습니다.')

        self.balance += amount
        self.deposit_num += 1       # 입금이 정상적으로 작동할 경우 +1 
        if self.deposit_num % 5 == 0:   # 5번 입금할 때 마다 이자를 지급 
            self.interest = self.balance * 0.01
            # :[repact] int() 함수 제거. (파이썬에서 소수점이하 버림이 됨)
            self.balance = self.balance + self.interest  

        # 입금 로그 기록 리스트에 추가
        # :[repact] 금액에 쉼표 포매팅 추가 (잔액은 이자계산을 고려해서 :.,0f 포매팅)
        log_message_deposit = f"입금액: {amount:,}원 | 일시: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
        self.deposit_list.append(log_message_deposit)

        print(f"입금이 완료되었습니다. [잔액 : {self.balance:,.0f}원]")

    # 출금 메서드
    def withdraw(self, amount):
        # 잔고보다 출금액이 클 경우 
        if amount > self.balance:
            raise ValueError('오류: 잔고 이상으로 출금할 수 없습니다.')

        self.balance -= amount

        # 출금 기록 저장
        log_message_withdraw = f"출금액: {amount:,}원 | 일시: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
        self.withdraw_list.append(log_message_withdraw)

        print(f"출금이 완료되었습니다. [잔액: {self.balance:,.0f}원]")

    # 입금 내역 출력 메서드
    def deposit_history(self):
        print(f"--- {self.name}님의 입금 내역 ---")
        # 입금한 내역이 없는 경우 
        if not self.deposit_list:
            print("입금내역이 없습니다.")
            return

        for i, history in enumerate(self.deposit_list, start=1):
            print(f"{i} {history}")

    # 출금 내역 출력 메서드
    def withdraw_history(self):
        print(f"--- {self.name}님의 출금 내역 ---")
        # 출금한 내역이 없는 경우 
        if not self.withdraw_list:
            print("출금내역이 없습니다.")
            return

        for i, history in enumerate(self.withdraw_list, start=1):
            print(f"{i} {history}")

    # 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.name}, 계좌번호: {self.account}, 잔고: {self.balance:,.0f}")
````

````python
# 계좌 3개 생성 
account1 = Account("일길동", 50000)
account2 = Account("이길동", 3000000)
account3 = Account("삼길동", 2000000)
````

````python
account_list = [account1, account2, account3]
````

````python
# 계좌 잔고가 100만원 이상인 고객의 정보 출력 

for i in account_list:
    if i.balance >= 1000000:
        i.display_info()
````

````python
# 입금-입금내역-출금내역 조회 시뮬레이션 
account1.deposit(10000)
account1.deposit(10000)
account1.deposit(10000)
account1.deposit(10000)
account1.deposit(10000)
account1.deposit(10000)
account1.deposit_history()
account1.withdraw_history()
````

````python
# 입금 에러 발생 시키기 - 1원 이하 입금

account1.deposit(0)
````

````python
# 출금 에러 발생 시키기 - 입금액 보다 많이 출금

account2.withdraw(100000000)
````

````python
# 출금-출금내역-입금내역 시뮬레이션 
account2.withdraw(100000)
account2.withdraw(100000)
account2.withdraw_history()
account2.deposit_history()
````

## 퀘스트 2) 간단한 자동 RPG 만들기

---

````python
import random

# 부모 클래스 Character 정의
class Character:
    def __init__(self, name, level, stamina, attack, defense):
        self.name = name
        self.level = level
        self.stamina = stamina
        self.attack = attack
        self.defense = defense

    # 캐릭터 생존 여부 확인 함수 
    def is_alive(self):
        return self.stamina > 0   # stamina가 0 이상이면 True, 아니면 False 리턴 

    # 공격 받았을 때 피해량 적용 함수 
    def take_damage(self, damage):
        actual_damage = damage - self.defense   # 실제 피해량 계산 

        # 최종 피해량이 0보다 클때,
        if actual_damage > 0:
            self.stamina -= actual_damage

            #[repact] 체력이 0 이하가 될 경우 0으로 값 수정 
            if self.stamina < 0 :
                self.stamina = 0
            print(f"{self.name}의 체력 : {self.stamina}")
        else:
            print(f"{self.name}의 방어력이 높아 데미지를 입지 않습니다.")

    #공격할 때 줄 데미지 적용 함수 
    def attack_target(self, target):
        if not self.is_alive():
            print(f"{self.name}은(는) 이미 쓰러져서 공격할 수 없습니다.")
            return

        damage = random.randint(1, self.attack)
        print(f"{self.name}이(가) {target.name}에게 {damage}만큼 공격했다...!")

        # 타겟의 take_damege 메서드를 실행히켜 데미지를 입힘
        target.take_damage(damage)


class Player(Character):
    def __init__(self, name):   # 닉네임 매개변수로 받음
        '''
        [Error_repo] super().__init__() 에 값을 안 넣고 호출해서
        TypeError: Character.__init__() missing 5 required positional arguments:
        'name', 'level', 'stamina', 'attack', and 'defense' 오류 발생
        '''
        super().__init__(name, 1, 100, 25, 5)
        self.exp = 0   # 플레이어 고유 특성

    def gain_experience(self, exps):
        self.exp += exps

    def level_up(self):
        if self.exp >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.exp = 0     # 레벨업 후 경험치 초기화



class Monster(Character):
    def __init__(self, name, level):
        '''
        [Error_repo] 랜덤값에 레벨을 곱할 때, self.level로 입력해서
        AttributeError: 'Monster' object has no attribute 'level' 오류 발생
        아직 self.level은 정의되지 않아서 발생한 오류
        '''
        super().__init__(name, level,
                         random.randint(10, 30) * level,
                         random.randint(5, 20) * level,
                         random.randint(1, 5) * level)
````

````python
# 플레이어와 몬스터 생성 

p1 = Player('Hong')
m1 = Monster('Tamguanori_1', 1)
m2 = Monster('Tamguanori_2', 5)
m3 = Monster('Tamguanori_2', 10)
````

````python
def battle(player, monster):
    print(f"{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():  # 플레이어와 몬스터의 체력이 0 이상일 때 반복 
        player.attack_target(monster)    # 플레이어가 먼저 공경

        if monster.is_alive():           # 플레이어 공격 이후 몬스터의 체력이 0 이상일 경우            
            monster.attack_target(player)

        print("-" * 30) # 턴 구분선

    if player.is_alive():
        player.gain_experience(monster.level * 20)
        player.level_up()

        print("전투 승리!")
        print(f"{monster.name}을 이겼다!")

    else:
        print("전투 패배...")
````

````python
# battle 함수 시뮬레이션 

battle(p1, m1)
````

````python
def main():
    monster_dict = {'슬라임':1, '고블린':2, '오크':3}
    player = Player(input("닉네임을 입력해주세요: "))
    for name, level in monster_dict.items():   # 딕셔니리의 요소를 차례로 반복 
        monster = Monster(name, level)
        print(f"야생의 {monster.name}(Lv.{monster.level})이(가) 나타났다!")

        battle(player, monster)

        if not player.is_alive():   # 플레이어가 사망하면 종료 
            break

# RPG 시뮬 시작 
main()
````
