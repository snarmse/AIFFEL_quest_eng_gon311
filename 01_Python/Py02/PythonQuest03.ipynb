## 문제 1. 최댓값과 최솟값 찾기 

```python
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())  
print("최댓값:", find_max())  
```

```python
def find_min_max(numbers):
    min_value = float('inf')    # 최소값을 찾기 위해 양의 무한대로 초기화. 어떤 숫자보다도 큰 수로 설정
    max_value = float('-inf')   # 최댓값을 찾기 위해 음의 무한대로 초기화. 어떤 숫자보다도 작은 수로 설정 

    def update_min_max(num):
        # 외부함수의 변수인 min_value, max_value 참조 
        nonlocal min_value
        nonlocal max_value

        if num < min_value :   # num이 min_value보다 작다면 min_value 변수에 num 값을 바인딩 
            min_value = num 
        if num > max_value :   # num이 max_value보다 크다면 max_value 변수에 num 값을 바인딩 
            max_value = num 

    # numbers 리스트의 모든 값을 순환하며 최댓값 최소값을 업데이트 
    for num in numbers : 
        update_min_max(num)

    # 최솟값을 반환하는 내부 함수: 
    def get_min():
        return min_value

    # 최댓값을 반환하는 내부 함수: 
    def get_max():
        return max_value 

    return get_min, get_max 
```

## 문제 2. 함수의 호출 횟수를 세는 데코레이터

지난 시간에 배운 데코레이터를 기억하시나요? 데코레이터도 내부함수를 반환하는 클로저를 활용한 문법입니다. 클로저를 활용하여 함수 호출 횟수를 출력하는 데코레이터를 만들어봅시다.

조건:  
+ 함수가 실행될 때 마다 함수 이름과 실행 횟수를 출력하는 데코레이터 `counter` 만들기  

힌트:  
+ 데코레이터의 내부 함수(클로저)는 실행횟수를 기억하기 위해 외부 함수의 변수를 참조 해야 합니다  
+ 함수의 이름은 `함수이름.__name__`으로 접근할 수 있습니다.

```python
# 함수 이름 및 실행 횟수 출력 하는 함수 정의 
def counter(func):
    count = 0        # 처음 counter 함수가 시작될 때 count 변수는 0으로 초기화 
    def wrapper():    # 외부 함수에서 인수를 가진 함수를 매개변수로 받을 경우, *agrs, **kwagrs 괄호 안에 추가 
        nonlocal count 
        # count ++      # [error] 자바와 달리 ++ 은 동작하지 않는다 
        count += 1      # 클로저 함수로 실행 할 때 마다 count 누적
        # 외부함수에서 매개변수로 받은 함수 실행 
        func()
        print(f"{func.__name__} 실행횟수 : {count}")
    return wrapper
```

```python
@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello() 
```
