## Q1 회문 판독 및 분류기 
### Q1-1) 회문 판독기

```python
# 회문 판독 함수 is_palindrome 정의

def is_palindrome(t):
    """ 문자열이 회문인지 여부 판변 

    Args : 
        t(str) : 판별할 문자열

    Returns : 
        bool : 회문이면 True, 비회문이면 False 
    
    """
    # 문자열의 절반만 돌며 양 끝 문자가 일치하는지 비교 
    for i in range(len(t)//2):    
        if t[i] != t[-i-1)]:      
            return False         # 한 쌍이라도 같지 않다면 비회문으로 분류 
    return True
```

```python
target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

### Q1-2) 회문 분류기 

```python
# 회문 분류 함수 palindrome_classifier 정의 


def palindrome_classifier(t):
    """ 입력된 단어 리스트에서 회문과 비회문을 분류하고 통계 반환

    Args : 
        t(list) : 검사할 단어들이 담긴 리스트 

    Returns : 
        tuple : (분류 디렉션, 회문 개수, 회문 비율)
    
    """
    
    pal_t = []                       # 회문 저장
    pal_f = []                       # 비회문 저장 
    for target in t:
        if is_palindrome(target) : 
            pal_t.append(target)     
        else : 
            pal_f.append(target)     

    # 통계 자료 계산 
    count = len(pal_t)               # 회문 총 개수  
    per = float(len(pal_t)/len(t))   # 전체 단어 중 회문 비율 

    dict_pal = {'회문O':pal_t, '회문X':pal_f}    # 결과를 딕셔너리로 구조화 

    return (dict_pal, count, per)
```

```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
```

## Q2 성적 분석기  
### Q2-1) 성적 데이터화 

```python
subjects = ["국어", "영어", "수학"]
students = ["유원영", "설지민", "신윤아", "장윤아"]
scores = [
    [94, 91, 89],
    [88, 96, 90], 
    [86, 88, 77],
    [98, 76, 92]
]

# 학생별 과목 점수 이중 딕셔너리 형태로 구조화 
scores_dict = {
    name: dict(zip(subjects, row)) 
    for name, row in zip(students, scores)
}

scores_dict
```

### Q2-2) 성적 분석기 

```python
# 과목별 평균 점수 딕셔너리로 구조화 
class_avg = {
    "국어" : sum(row[0] for row in scores) / len(students), 
    "영어" : sum(row[1] for row in scores) / len(students), 
    "수학" : sum(row[2] for row in scores) / len(students)
}

# 전체 과목 평균 점수 
class_total_avg = sum(sum(scores, [])) / (len(students)*len(subjects))

# 성적 분석 함수 score_analysis 정의 
def score_analysis(name):
    """ 지정된 학생의 과목별 점수를 학급 평균과 비교하여 분석 결과 출럭 

    Args : 
        name(str) : 성적 분석할 학생 이름 
    """ 
    # 학생 개인 데이터 및 총점, 평균 계산 
    student_score = scores_dict[name]
    student_total = sum(student_score.values())
    student_avg = student_total / len(student_score)
    
    print(f"[{name} 성적 분석]")
    # 과목별 점수 및 학급 평균과의 편차 
    print(f"국어 : {student_score['국어']:.2f}점, 학급 평균({class_avg['국어']:.2f}점) 대비 {student_score['국어'] - class_avg['국어']:+.2f}")
    print(f"영어 : {student_score['영어']:.2f}점, 학급 평균({class_avg['영어']:.2f}점) 대비 {student_score['영어'] - class_avg['영어']:+.2f}")
    print(f"수학 : {student_score['수학']:.2f}점, 학급 평균({class_avg['수학']:.2f}점) 대비 {student_score['수학'] - class_avg['수학']:+.2f}")
    print("")
    # 전과목 평균 비교 결과 출력 
    print(f"전과목 평균: {student_avg:.2f}점, 학급 평균({class_total_avg:.2f}점) 대비 {student_avg - class_total_avg:+.2f}")
```

```python
score_analysis("설지민")
```
