# 📊 성적 분석기 파트 리팩토링 (Quest C 02 - Q.2-2)


```python
subjects = ["국어", "영어", "수학"]
students = ["유원영", "설지민", "신윤아", "장윤아"]
scores = [
    [94, 91, 89],
    [88, 96, 90],
    [86, 88, 77],
    [98, 76, 92]
]

scores_dict = {
    name: dict(zip(subjects, row)) for name, row in zip(students, scores)
}

print(scores_dict)
```

```python
'''기존 작성 코드
# class_avg = {
#     "국어" : sum(row[0] for row in scores) / len(students),
#     "영어" : sum(row[1] for row in scores) / len(students),
#     "수학" : sum(row[2] for row in scores) / len(students)
# }

#
# class_total_avg = sum(sum(scores, [])) / (len(students)*len(subjects))
'''

# 과목별 학급 평균 점수 
# class_avg = {sub: (sub의 점수를 다 더해서 나누기) for sub in subjects}
    # subjects 를 돌 때 몇 번째 과목인지 번호를 같이 뽑아서 row[i]로 접근 => enumerate 사용
    # class_avg = {
    #     sub: sum(row[i] for row in scores) / len(students)
    #     for i, sub in enumerate(subjects)
    # }

# scores_dict 사용
class_avgs = {
    sub: sum(scores_dict[st_name][sub] for st_name in students) / len(students)
    for sub in subjects
}


# 학생 성적 분석 함수 정의
def score_analysis(name):
    # 학생 이름(key)를 이용해서 아이템 뽑아내기
    student = scores_dict[name]

    print(f"{name} 성적 분석")

    total_student = 0   # 학생 점수 총합
    total_class = 0     # 반 점수 총합

    # 과목별 분석 루프
    for sub in subjects:
        student_score = student[sub]         # 학생의 과목별 성적
        sub_avg = class_avgs[sub]            # 해당 과목 평균

        diff = student_score - sub_avg

        print(f"{sub} : {student_score}점, "
              f"학급 평균({sub_avg:.2f}점) 대비 {diff:+.2f}")

        total_student += student_score
        total_class += sub_avg

    num_subjects = len(subjects)                   # 과목수의 증감에 대응 가능 
    student_avg = total_student / num_subjects
    class_avg = total_class / num_subjects
    total_diff = student_avg - class_avg

    print(f"\n전과목 평균 : {student_avg:.2f}점, "
          f"학급 평균({class_avg:.2f}점) 대비 {total_diff:+.2f}")


score_analysis("설지민")
```
