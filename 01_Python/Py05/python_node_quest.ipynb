## Q1) 사각형 넓이 구하기 
  
 
[목표] : 사각형, 평행사변형, 사다리꼴 넓이를 구하는 메서드를 가지는 클래스 구현 

---

````python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    # 직사각형 메서드 
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 메서드
    def par(self):
        lower, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = lower * height
        result = '평행사변형 넓이는 : ' + str(area)
        return result

    # 사다리꼴 메서드
    def trape(self):
        lower, upper, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))
        area = (lower + upper) * height * 0.5
        result = "사다리꼴 넓이는 : " + str(area)
        return result
````

````python
a = Square()
````

````python
a.rect()
````

## Q2) 키오스크 만들기 


[목표] : 음료 주문, 결제, 주문표 출력을 실행하는 메서드를 포함하는 키오스크 클래스 구현 

---

````python
import datetime

# 데코레이션 함수 : 외곽선 그리는 함수 정의
def write_line(func):
    def outline(self):
        # 외곽 정의
        outline = '⟝' + '-' * 30 + '⟞'
        side_line = '|' + ' ' * 31 + '|'

        # 외곽 출력
        print(outline)
        for i in range(5):
            print(side_line)
        func(self)
        for i in range(5):
            print(side_line)
        print(outline)
    return outline

# 키오스크 클래스 정의
class Kiosk:
    # 생성자 함수 정의
    def __init__(self):
        # 유지보수를 위해 다차원 리스트로 menu 속성 정의
        self.menu =[
            ['americano', 2000],
            ['latte', 3000],
            ['mocha', 3000],
            ['yuza_tea', 2500],
            ['green_tea', 2500],
            ['choco_latte', 3000],
        ]

    # 메뉴 출력 메서드
    def menu_print(self):
        print("---- 메뉴판 ----")
        for i, menu_info in enumerate(self.menu, start=1): # 메뉴 개수만큼 반복
            print(f"{i} {menu_info[0]} : {menu_info[1]}")   # 메뉴와 가격 출력

    # 주문 메서드
    def menu_select(self):
        print()
        self.order_menu = []   # 주문 메뉴 리스트 초기화
        self.order_price = []  # 주문 가격 리스트 초기화
        self.price_sum = 0     # 총 합계 초기화

        is_first_order = True   # 첫 주문인지를 확인

        while True :
            print()
            if not is_first_order:           # 첫 주문이 아닌 경우 추가 주문 여부 안내 문구 출력 
                print("추가 주문을 하시겠습니까?")

            n = int(input("음료 번호를 입력하세요 (지불 및 종료는 0번): "))

            # 1. 종료 조건을 만족할 때 (0 입력)
            if n == 0:
                if not self.order_menu :
                    print("주문한 내역이 없습니다")
                    continue
                print("\n====================")
                print("주문이 완료되었습니다.")
                print(f"주문 내역: {self.order_menu}")
                print(f"총 결제 금액: {self.price_sum}원")
                print("=======================")
                break

            # 2. 메뉴판에 있는 번호일 때 (정상 주문 처리)
            elif 1 <= n <= len(self.menu):
                # 음료 온도 물어보기
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else :
                        print("1과 2 중 하나를 입력해주세요.\n")

                # 데이터 저장 및 합계 누적
                selected_menu = self.menu[n-1][0]
                selected_price = self.menu[n-1][1]

                full_menu_name = f"{self.temp} {selected_menu}"
                self.order_menu.append(full_menu_name)
                self.order_price.append(selected_price)
                self.price_sum += selected_price

                print(f"\n[담긴 상품] {full_menu_name} ({selected_price}원)")
                print(f"[현재 합계] {self.price_sum}원")

                # 첫 주문이 완료된 경우, 추가주문 문구가 뜰 수 있도록 속성 변경
                is_first_order = False

            # 3. 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해주세요.")


    # 지불 메서드
    def pay(self):
        # 합계 금액 출력
        print(f"총 합계 금액은 {self.price_sum}원 입니다 ")

        self.pay_method = 0   # 지불 방법 초기화
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input("현금 결제는 1을, 카드 결제는 2를 입력하세요 : "))

            if self.pay_method == 1:
                print("직원을 호출하겠습니다.\n")
            elif self.pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.\n")
            else:
                print("1과 2 중 하나를 입력하세요.\n")

    # 주문서 출력 메서드
    @write_line
    def table(self):
        # 주문 일시 출력
        print("주문 일자 : ", datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), "\n")

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], " : ", self.order_price[i])

        # 합계 금액 출력
        print('합계 금액: ', self.price_sum)
````

````python
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
````
